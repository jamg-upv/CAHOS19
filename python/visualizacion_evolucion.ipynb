#**Empezamos importando el csv orignial**#
llamado "Capacidad_Asistencial_Historico_gobes.csv", convirtiendo las columnas que tocan en fechas



```python
import pandas as pd
import seaborn as sns
import numpy as np
from datetime import datetime, timedelta

direccion0="https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/inputtables/Capacidad_Asistencial_Historico_gobes.csv"
capacidad_asistencial=pd.read_csv(direccion0, delimiter=';', encoding='latin_1')
capacidad_asistencial['Fecha'] = pd.to_datetime(capacidad_asistencial['Fecha'], format="%d/%m/%Y")

byprovincia=capacidad_asistencial.groupby(['Provincia'], as_index=False).agg('mean')
byprovincia
```

```python
#ordena el archivo primero por fechas y luego por código provincial(cap_Asis) Y ccaa(cap_asis00)
cap_asis = capacidad_asistencial.sort_values(['Fecha', 'Cod_Provincia'])
cap_asis00=capacidad_asistencial.sort_values(['Fecha', 'CCAA'])
```

```python
#Lo que hacemos es agrupar según provincia (cap_asis1) y por CCAA (cap_asis01), y luego por fecha para que haga la suma de las de: TOTAL_CAMAS, OCUPADAS_COVID19, OCUPADAS_NO_COVID19
cap_asis1=cap_asis.groupby(['Cod_Provincia', 'Fecha'])['TOTAL_CAMAS', 'OCUPADAS_COVID19', 'OCUPADAS_NO_COVID19'].apply(sum)
cap_asis01=cap_asis00.groupby(['CCAA', 'Fecha'])['TOTAL_CAMAS', 'OCUPADAS_COVID19', 'OCUPADAS_NO_COVID19'].apply(sum)

cap_asis01
```

```python
#Ahora lo que buscamos es dejar solo la comulma de OCUPADAS_COVID19 en funcion de la provincia y las fechas
cap_asis2=cap_asis1.drop(['TOTAL_CAMAS', 'OCUPADAS_NO_COVID19'], axis=1)
cap_asis02=cap_asis01.drop(['TOTAL_CAMAS', 'OCUPADAS_NO_COVID19'], axis=1)
cap_asis02
```

```python
#Generamos un archivo CSV con la nueva tabla creada con los nuevos agrupamientos

cap_asis2.to_csv('provincias_camas_totuci_covid_nocovid.csv')
cap_asis02.to_csv('CCAA_camas_totuci_covid_nocovid.csv')
```

#**Importamos nuevo csv reorganizado y estructurado (Cod_Provincia y CCAA)**#
llamado: "provincias_camas_totuci_covid_nocovid.csv" por código y "CCAA_camas_totuci_covid_nocovid.csv" por CCAA de los outputtable

Aquí lo que haremos será la puesta a punto para meterlo directamente en el flourish (transpondremos filas y columnas)

```python
direccion1="https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/outputtables/provincias_camas_totuci_covid_nocovid.csv"
direccion01="https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/outputtables/CCAA_camas_totuci_covid_nocovid.csv"
provincias_camas=pd.read_csv(direccion1, delimiter=',', encoding='latin_1')
CCAA_camas_COVID=pd.read_csv(direccion01, delimiter=',')
CCAA_camas_COVID
```

```python
#Intento trasponer las columnas de fecha y 'OCUPADAS_COVID' por columnas para poder hacer el gráfico

provincias_camas1=provincias_camas.pivot(index='Cod_Provincia', columns='Fecha', values='OCUPADAS_COVID19')
CCAA_camas_COVID1=CCAA_camas_COVID.pivot(index='CCAA', columns='Fecha', values='OCUPADAS_COVID19')
CCAA_camas_COVID1
```

```python
#Exportamos la nueva tabla para representar en el florish
provincias_camas1.to_csv('provincias_camas_covid.csv')
CCAA_camas_COVID1.to_csv('CCAA_camas_COVID1.csv')
```

#**Realizamos archivo para tener camas con covid/camas totales**#

```python
cap_asis3=cap_asis01.drop(['OCUPADAS_NO_COVID19'], axis=1) #Eliminamos la que no nos va a ser útil en esta
PORCENTAJE_COVID=(cap_asis3['OCUPADAS_COVID19']/cap_asis3['TOTAL_CAMAS'])*100 #Calculamos los % de camas con covid sobre el total
cap_asis3['PORCENTAJE_COVID']=PORCENTAJE_COVID #Lo metemos en una columna del Data Frame
cap_asis31=cap_asis3.drop(['OCUPADAS_COVID19', 'TOTAL_CAMAS'], axis=1)#Eliminamos el resto de columnas dejando solo la del %
cap_asis31=cap_asis31.round({'PORCENTAJE_COVID':2}) #Hacemos que el % solo tenga dos decimales

cap_asis31
```

```python
#Exportamos a un csv nuevo para poder trasponer las columnas y poder hacer la bar race
cap_asis31.to_csv('porcentaje_covid_CCAA.csv')
#Importamos nuevamente el csv
direccion3="https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/outputtables/porcentaje_covid_CCAA.csv"
porcentaje_covid_CCAA=pd.read_csv(direccion3, delimiter=',')
#Trasponemos las columnas a filas
porcentaje_camascovid_CCAA=porcentaje_covid_CCAA.pivot(index='CCAA', columns='Fecha', values='PORCENTAJE_COVID')
#Exportamos csv final para la bar race
porcentaje_camascovid_CCAA.to_csv('porcentaje_camascovid_CCAA.csv')
```

#**Realizamos archivo para tener camas ocupadas/camas totales**#

```python
cap_asis4=cap_asis01
CAMAS_TOTALES_OCUPADAS=(cap_asis4['OCUPADAS_COVID19']+cap_asis4['OCUPADAS_NO_COVID19'])
cap_asis4['CAMAS_TOTALES_OCUPADAS']=CAMAS_TOTALES_OCUPADAS
PORCENTAJE_CAMAS_TOTALES_OCUPADAS=(cap_asis4['CAMAS_TOTALES_OCUPADAS']/cap_asis4['TOTAL_CAMAS'])*100
cap_asis4['PORCENTAJE_CAMAS_TOTALES_OCUPADAS']=PORCENTAJE_CAMAS_TOTALES_OCUPADAS
cap_asis4=cap_asis4.round({'PORCENTAJE_CAMAS_TOTALES_OCUPADAS':2})
cap_asis41=cap_asis4.drop(['TOTAL_CAMAS', 'OCUPADAS_COVID19','OCUPADAS_NO_COVID19', 'CAMAS_TOTALES_OCUPADAS'], axis=1)

cap_asis41.to_csv('porcentaje_camas_totales_ocupadas.csv')
```

```python

direccion4="https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/outputtables/porcentaje_camas_totales_ocupadas.csv"
porcentaje_camas_totales_ocupadas=pd.read_csv(direccion4, ',')
porcentaje_camas_totales_ocupadas_barrace=porcentaje_camas_totales_ocupadas.pivot(index='CCAA',columns='Fecha', values='PORCENTAJE_CAMAS_TOTALES_OCUPADAS')
porcentaje_camas_totales_ocupadas_barrace
porcentaje_camas_totales_ocupadas_barrace.to_csv('porcentaje_camas_totales_ocupadas_barrace.csv')
```
