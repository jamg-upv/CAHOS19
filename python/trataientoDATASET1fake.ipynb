```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

direccion0='https://raw.githubusercontent.com/jamg-upv/CAHOS19/main/inputtables/dataset1fake1.csv'
df=pd.read_csv(direccion0, delimiter=';')

df['parasumar']=1 #esta columna se usar√° para hacer los recuentos totales del dataset2

df
```

```python
#Eliminamos las columnas que no nos sirven para cada suma
ingresoshosp=df.drop(['Id', 'fechaingreuci', 'fechaaltauci', 'fechaaltahosp', 'fechadefunc', 'fechatraslado', 'altadeftraslado','ComorbidityIndex'], axis=1)
ingresosuci=df.drop(['Id', 'fechaingrehosp', 'fechaaltauci','fechaaltahosp', 'fechadefunc', 'fechatraslado', 'altadeftraslado','ComorbidityIndex'], axis=1)
altasuci=df.drop(['Id', 'fechaingrehosp', 'fechaingreuci','fechaaltahosp', 'fechadefunc', 'fechatraslado', 'altadeftraslado','ComorbidityIndex'], axis=1)
altashosp=df.drop(['Id', 'fechaingrehosp', 'fechaingreuci','fechaaltauci', 'fechadefunc', 'fechatraslado', 'altadeftraslado','ComorbidityIndex'], axis=1)
defunc=df.drop(['Id', 'fechaingrehosp', 'fechaingreuci','fechaaltauci', 'fechaaltahosp', 'fechatraslado', 'altadeftraslado','ComorbidityIndex'], axis=1)
traslado=df.drop(['Id', 'fechaingrehosp', 'fechaingreuci','fechaaltauci', 'fechaaltahosp', 'fechadefunc', 'altadeftraslado','ComorbidityIndex'], axis=1)
adt=df.drop(['Id', 'fechaingrehosp', 'fechaingreuci','fechaaltauci', 'ComorbidityIndex'], axis=1)

#arreglar defunc eliminando las filas nan de defunc y traslado
defunc=defunc.dropna()
traslado=traslado.dropna()
```

```python
#para transformar las columnas de fecha en fechas (lo hacemos despu√©s para poder eliminar correctamente los nan)
df['fechaingrehosp']=pd.to_datetime(df['fechaingrehosp'], format='%Y-%m-%d')
df['fechaingreuci']=pd.to_datetime(df['fechaingreuci'], format='%Y-%m-%d')
df['fechaaltauci']=pd.to_datetime(df['fechaaltauci'], format='%Y-%m-%d')
df['fechadefunc']=pd.to_datetime(df['fechadefunc'], format='%Y-%m-%d')
df['fechatraslado']=pd.to_datetime(df['fechatraslado'], format='%Y-%m-%d')
```

```python
#Agrupamos por las fechas, sexo, edad y hospital, y sumamos el total que hay de cada categor√≠a
ingresoshosp=ingresoshosp.groupby(['fechaingrehosp', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
ingresosuci=ingresosuci.groupby(['fechaingreuci', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
altasuci=altasuci.groupby(['fechaaltauci', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
altashosp=altashosp.groupby(['fechaaltahosp', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
defunc=defunc.groupby(['fechadefunc', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
traslado=traslado.groupby(['fechatraslado', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
adt=adt.groupby(['fechaaltahosp', 'fechadefunc', 'fechatraslado','altadeftraslado', 'sexo', 'edad', 'CNH'])['parasumar'].sum()
```

```python
ingresoshosp.to_csv('ingresoshosp.csv')
ingresosuci.to_csv('ingresosuci.csv')
altasuci.to_csv('altasuci.csv')
altashosp.to_csv('altashosp.csv')
defunc.to_csv('defunc.csv')
traslado.to_csv('traslado.csv')
adt.to_csv('adt.csv')
```
